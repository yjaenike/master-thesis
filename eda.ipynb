### Breakdown of one patients data

```python
df_path = '../icp-flask/data/DB_UKE_60_final_df.parquet'
print("Loading from Path: '{}' ...".format(df_path), end='')
df_icu_data = pd.read_parquet(df_path)
print(" Done. ✅\n")

pat_id = 484096897
one_patient_data = df_icu_data[df_icu_data["Pat_ID"] == pat_id]
if pat_id not in df_icu_data['Pat_ID'].to_list(): print("\033[91m\033[1m!!! Not a patient id !!!\033[0m")
```

```python
import ipywidgets as widgets

# remove columns
only_float_columns = []
for column_name in df_icu_data.columns:
    series = df_icu_data[column_name]
    if series.dtype == "float64" and series.name != "rel_time":
        only_float_columns.append(series.name)   
       
# create tuple list
drop_down_options = []
for col_name in only_float_columns:
    column_name_tuple = (col_name.replace("_"," "), col_name)
    drop_down_options.append(column_name_tuple)

# Drop down menu
select = 'AF_Vital'
drop_down = widgets.Dropdown(
    options=drop_down_options,
    value="AF_Vital",
    description='Series:',
)

def dropdown_handler(change):
    global select
    
    Javascript('Jupyter.execute_cell_range(IPython.notebook.get_selected_index()+1, IPython.notebook.get_selected_index()+2)')
    
    select = change.new  # This line isn't working
    
drop_down.observe(dropdown_handler, names='value')
print("Select the series to show in the plot here:")
display(drop_down)
```

```python
sns.set(rc={"figure.figsize":(15, 5)}) #width=3, #height=4
fig, ax = plt.subplots()
sns.scatterplot(data = one_patient_data, x= "rel_time",y=select, ax=ax).set(title="'{}' over time".format(select.replace("_", " ")), xlabel='Relative Time')
ax.set_xlim(-10,one_patient_data["rel_time"].to_list()[-1])
plt.show()
```

### ICU bed occupancie length

```python
df_path = '../icp-flask/data/DB_UKE_60_final_df.parquet'
print("Loading from Path: '{}' ...".format(df_path), end='')
df_icu_data = pd.read_parquet(df_path)
print(" Done. ✅\n")

hours_per_patient = df_icu_data.groupby("Pat_ID")[["rel_time"]].count().apply(lambda x: x/24)
```

```python
sns.set(style="whitegrid")
sns.set_context('poster', font_scale = 1)

ax = sns.histplot(hours_per_patient, stat='count')
ax.set(xlabel='time on ICU [days]', ylabel='number of patients')
```
